```sage
var("y, x")
y1 = 2*x
y2 = sin(pi*x)

C1 = plot(y1, (x, 0, 1/2), color="blue", legend_label="$y = 2x$", fill=y2)
C2 = plot(y2, (x, 0, 1/2), color="green", legend_label=r"$y = \sin(\pi x)$")

show(C1 + C2, aspect_ratio=1/2, legend_loc="right")
```

```sage
# Plot the region
var('y, x')

y1 = 0
y2 = x
y3 = -2*x + 6

C1 = plot(y1, (x, 0, 3.5), color='blue', legend_label='$y = 2x$', fill= [y2, y3])
C2 = plot(y2, (x, 0, 3.5) ,color = 'green', legend_label = r'$y = \sin(\pi x)$', fill=[y1, y3])
C3 = plot(y3, (x, 0, 3.5) ,color = 'purple', legend_label ='$y = -2x + 6$', fill = [y2, y3])

show(C1+C2+C3, aspect_ratio = 1, legend_loc="upper right")
```

```sage
var('y, x')

y_top(x) = e^x
y_bottom(x) = 0

C1 = plot(y_top, (x, 0, 2), color='green', legend_label=r'$y = e^x$')
C2 = plot(y_bottom, (x, 0, 2), color='blue', legend_label='$y = 0$', fill=y_top)
L1 = line([(0, 0), (0, 1)], color='purple', legend_label='$x = 0$', linestyle='--')
L2 = line([(2, 0), (2, e^2)], color='red', legend_label='$x = 2$', linestyle='--')

show(C1 + C2 + L1 + L2, legend_loc="upper left", ymin=0, ymax=8)
```
