```sage
"""
Let the following coordinates:
A = (-73, 41)
B = (-90, 30)
C = (-80, 26)

x_to_origin = B - A = (-90+73, 30 - 41) = (-17, -11)
y_to_origin = C - B = (-80 + 90, 26 - 30) = (10, -4)
z_to_origin = A - C = (7, 15)

"""

# Plotting the vectors x, y, z
x_vec = arrow((-90, 30), (-73, 41), color = 'blue') # B -> A
y_vec = arrow((-80, 26), (-90, 30), color = 'red') # C -> B
z_vec = arrow((-73, 41), (-80, 26), color = 'green')

x_vec + y_vec + z_vec
```

```sage
# Calculating the Magnitude of each one of the vectors:

ba = vector([-17, -11])
cb = vector([10, -4]) 
ac = vector([7, 15])
ba_magnitude = ba.norm()
cb_magnitude = cb.norm()
ac_magnitude = ac.norm()

print(f"Magnitude BA: {ba_magnitude}, Magnitude CB: {cb_magnitude}, Magnitude AC: {ac_magnitude}")
```
